## Handling Missing Values

```python
zomato_data_cleaned['address'].fillna('Unknown', inplace=True)
```

```python
zomato_data_cleaned.drop('zipcode', axis=1, inplace=True)
```

```python
zomato_data_cleaned['cuisines'].fillna('Unknown', inplace=True)
```

```python
zomato_data_cleaned['timings'].fillna('Not Available', inplace=True)
```

## Competitive Analysis

```python
# Step 1: Create a combined category identifier
zomato_data_cleaned['category_identifier'] = zomato_data_cleaned['region'] + "_" + zomato_data_cleaned['cuisines'] + "_" + zomato_data_cleaned['price_range'].astype(str)

# Step 2: Calculate average ratings for each category
average_ratings = zomato_data_cleaned.groupby('category_identifier')['aggregate_rating'].mean()

# Step 3: Identify top restaurants in each category
# Define a threshold for being a top restaurant, e.g., having a rating above the category average
top_restaurants = zomato_data_cleaned[zomato_data_cleaned.apply(lambda x: x['aggregate_rating'] > average_ratings[x['category_identifier']], axis=1)]

# Display top restaurants
print(top_restaurants[['name', 'region', 'cuisines', 'price_range', 'aggregate_rating']])
```

```python
# For this example, let's consider 'top_restaurants' as your DataFrame of major competitors

# Analyze Ratings
rating_analysis = top_restaurants.groupby('category_identifier')['aggregate_rating'].agg(['mean', 'min', 'max'])

# Analyze Popularity
popularity_analysis = top_restaurants.groupby('category_identifier')['votes'].agg(['mean', 'sum'])

# Combine the Analyses
combined_analysis = pd.concat([rating_analysis, popularity_analysis], axis=1)

# Interpret the Results
print(combined_analysis)
```

## Market Gap Analysis

```python
# Step 1: Analyze Cuisine Representation
# Splitting cuisines and counting their occurrences
zomato_data_cleaned['individual_cuisines'] = zomato_data_cleaned['cuisines'].str.split(', ')
cuisines_exploded = zomato_data_cleaned.explode('individual_cuisines')
cuisine_counts = cuisines_exploded.groupby(['region', 'individual_cuisines']).size().reset_index(name='counts')

# Identifying underrepresented cuisines in each region
# You can adjust the threshold as per your requirement
threshold_cuisines = cuisine_counts['counts'].quantile(0.25)  # For example, lower quartile
underrepresented_cuisines = cuisine_counts[cuisine_counts['counts'] < threshold_cuisines]

# Step 2: Assess Price Range Distribution
# Counting the number of restaurants in each price range by region
price_range_distribution = zomato_data_cleaned.groupby(['region', 'price_range']).size().reset_index(name='counts')

# Identifying potential price range gaps
# You can adjust the threshold as per your requirement
threshold_price = price_range_distribution['counts'].quantile(0.25)  # For example, lower quartile
potential_price_gaps = price_range_distribution[price_range_distribution['counts'] < threshold_price]

# Output the results
print("Underrepresented Cuisines in Each Region:\n", underrepresented_cuisines)
print("\nPotential Price Range Gaps in Each Region:\n", potential_price_gaps)
```

```python
# Assuming 'underrep_cuisines' is a DataFrame containing the underrepresented cuisines data
# Group by 'individual_cuisines' and sum the 'count' across all regions
aggregated_cuisines = underrep_cuisines.groupby('individual_cuisines')['count'].sum().reset_index()

# Sort the cuisines by count in descending order and select the top N
top_cuisines = aggregated_cuisines.sort_values(by='count', ascending=False).head(10)

# Plot the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='individual_cuisines', y='count', data=top_cuisines)
plt.title('Top Underrepresented Cuisines Across All Regions')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
